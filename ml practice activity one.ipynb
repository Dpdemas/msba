```python
import urllib.request
from urllib.request import urlopen
```

```python
# group members
items = ["Joe", "Dom"]
for item in items:
  print(f"- {item}")

  #something we enjoy

image = [
      "https://assets-prd.ignimgs.com/2022/04/19/wow-dragonflight-button-1-1650398895381.jpg",
      "https://im4.ezgif.com/tmp/ezgif-4-8ab2aab33f.jpg"
  ]

name = [
    "Joe",
    "Dom"
]

number_of_img = len(image)
#print(number_of_img)

fig, axes = plt.subplots(1, number_of_img, figsize=(15, 5))

for ax, image, name in zip(axes, image, name):
  with urlopen(image) as url:
    img = mpimg.imread(url, format='jpg')
    ax.imshow(img)
    ax.axis('off')
    ax.set_title(name)

plt.tight_layout()
plt.show()
```
